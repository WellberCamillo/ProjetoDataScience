```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set_theme(style="whitegrid")

df = pd.read_csv('Datasate/IceCreamData.csv')
df = round(df.loc[0:11])
```

```python
df.loc[0:11]
```

```python
result = df.loc[0:11].sort_values('Temperature', ascending = False)
result
```

```python
celsius = (df.loc[0:11]["Temperature"]- 32) * 5/9
celsius['Temperature'] = celsius
```

```python
plt.plot(result['Temperature'], result['Revenue'], 'r.--',label = 'US$')
plt.title('Ice Crem Truck')
plt.ylabel('Revenue US$')
plt.xlabel('Temperature (Celsius)')        
plt.legend()

plt.savefig('IceCreamData.pdf')

plt.show()
```
